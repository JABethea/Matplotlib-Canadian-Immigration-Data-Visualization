```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
df_can.rename(columns={'OdName' : 'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
#Index immigration data from Japan to Canada
print(df_can.loc['Japan'])

#alternate method
print(df_can.iloc[87])
print(df_can[df_can.index == 'Japan'].T.squeeze())
```

```python
#For a specific year
print(df_can.loc['Japan', 2013])

#alternate method
print(df_can.iloc[87, 36])
```

```python
#Now for multiple years, 1990 to 1995
print(df_can.loc['Japan', [1990, 1991, 1992, 1993, 1994, 1995]])

#alternate method
print(df_can.iloc[87, [13,14,15,16,17,18]])
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
#Filtering based on a criteria
#1. create a condition boolean series
condition = df_can['Continent'] == 'Asia'
print(condition)

#2. pass this condition into the dataframe
df_can[condition]
```

```python
inplace = True 
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose()

print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,8))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
